```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fun(x):
    return np.sin(4.8*np.pi*x)

def dfun(x):
    return 4.8*np.pi*np.cos(4.8*np.pi*x)
    
x0 = 0.2
dfx0_truth = dfun(x0)

def forward_difference(x, f, h):
    return (f(x+h) - f(x)) / h

hz = np.power(2., -np.arange(5,31))
errz = np.zeros(np.size(hz))
for i in range(len(hz)):
    
    h = hz[i]
    dfx0_approx = forward_difference(x0, fun, h)
    errz[i] = np.fabs(dfx0_truth - dfx0_approx) / np.fabs(dfx0_truth)
```

```python
plt.figure()
plt.loglog(hz, errz, 'bo-')
plt.grid(True)
plt.xlabel('h')
plt.ylabel('relative error')
plt.title('Fwd diff error: d/dx sin(4.8 pi x) at x=0.2')
plt.show()
```
